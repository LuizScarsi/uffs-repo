Como calcular o valor de um produto ap√≥s um desconto?

Por exemplo, um produto de R\$40 com 20% de desconto sairia por R\$32.

``valor = 40``

``desconto = 20``

O desconto √© informado no valor percentual, portanto 20% = 20/100 = 0.02.

O valor descontado √© igual ao valor multiplicado pelo desconto, ou seja, 40 x 0.02 = 8
O valor final √© igual o valor inicial menos o valor do desconto, ou seja, 40 - 8 = 32.

Como seria um programa? Vejamos a seguir:

````python
# Programa que computa o valor de um produto com desconto
# ENTRADA
valor = input('Digite o valor do produto: ')
valor = float(valor)
desconto = float(input('Digite o desconto, em valores percentuais, para o produto: '))
# PROCESSAMENTO
valor_final = valor - (valor * desconto / 100)
# SA√çDA
print(f'O valor final do produto, ap√≥s o desconto de {desconto}%, √© igual a R${valor_final}')
````

Voc√™ gosta de matem√°tica? N√£o? üòß

Matem√°tica √© muito legal pois nos d√° atalhos para alguns c√°lculos.

Vejamos:

Sabemos que se multiplicarmos um n√∫mero por 1, o resultado √© o mesmo: 200 *1 √© igual a 200

Sabemos que 0,01 √© 1% de qualquer n√∫mero. Ent√£o se subtrairmos 0,01 de 1 teremos 0,99.

Agora se multiplicarmos 0,99 por um n√∫mero qualquer, n√≥s iremos descontar 1% do valor.

M√°gica? N√£o matem√°tica ü•∞

````python
# Programa que computa o valor de um produto com desconto
# ENTRADA
valor = input('Digite o valor do produto: ')
valor = float(valor)
desconto = float(input('Digite o desconto, em valores percentuais, para o produto: '))
# PROCESSAMENTO - Outra forma de fazer o mesmo c√°lculo
valor_final = valor * (1 - desconto/100)
# SA√çDA
print(f'O valor final do produto, ap√≥s o desconto de {desconto}%, √© igual a R${valor_final}')
````

**6.2** -  Fa√ßa um programa que solicite o custo de produ√ß√£o de uma mercadoria e um valor percentual de lucro que ser√° acrescido ao valor do produto para a venda.

Imprimira o valor final do produto (valor de venda).

Desafio: compute e imprima tamb√©m o valor m√°ximo de desconto (em percentual) que a loja pode aplicar na mercadoria para que n√£o tenha preju√≠zo na venda.

O valor final do produto, ap√≥s desconto, deve ser maior ou igual ao custo de produ√ß√£o para que a empresa n√£o tenha preju√≠zo na venda.

````python
# Entrada
custo_producao = int(input('Informe o custo de produ√ß√£o: '))
lucro = int(input('Informe o percentual de lucro acrescentado ao valor: '))
# Processamento
valor_final = custo_producao * (1 + lucro/100)
print(f'O valor final √© {valor_final}')

desconto = int(input('Insira o percentual do desconto: '))
if desconto > lucro:
  print('Voc√™ vai falir')
else:
  valor_com_desconto = valor_final * (1 - desconto/100)
  print(f'O valor com desconto √© = {valor_com_desconto}')
````
