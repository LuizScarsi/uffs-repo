Uma variável possui um nome (identificador) e é associada a um tipo de dado.

````python
mensagem = "Hello Python world!\n!!!!"
print(mensagem)
````

````python
a = 1
b = 5
print(a)
print(b)
````

Regras para definição de nomes (identificadores) de variáveis:


1.   Um identificador de variável deve começar com letras ou com o símbolo underscore (_);
2.   Não é permitido começar com números ou caracteres especiais ( *, -, #, &, entre outros);
3.   Em Python o padrão mais utilizado é o uso do underscore (_) para separar as palavras de um nome composto ou longo (nome_composto) pois um nome não pode conter espaços;
4.   Letras maiúsculas e minúsculas são reconhecidas como caracteres diferentes; em Python o padrão mais utilizado é o uso de apenas caracteres mínusculos para nomear variáveis;
5.   O ideal é que o identificador de uma variável seja um nome significativo dentro do contexto do programa.


````python
_var = 1
Var = 2
x = 3
v_total = _var + Var + x
print(v_total)
````

Uma ``variável`` assim se denomina porque seu conteúdo ou valor pode variar durante a execução do programa.

Variáveis podem receber valores fixos (atribuídos no código-fonte) bem como valores provenientes de entradas do usuário (inputs) ou resultados de outras operações e cálculos.

O sinal de igual ``=`` é chamado de **operador de atribuição** e serve para atribuir um novo valor a uma variável. Lê-se como "**recebe**".

```variavel = valor ou expressão```

Note que, embora possa assumir diferentes valores no decorrer de um programa, uma variável somente pode armazenar um único valor a cada instante. Isto é, ao receber um novo valor, o valor anterior é sobrescrito (portanto, perdido).

````python
a = 2
print(a)
a = 3  # valor 2 se perdeu
print(a)
````

Os tipos básicos de dados do Python são:

*  Números inteiros (int): -14, -2, 0, 1, 100, 401233

A linguagem Python é dinamicamente tipada, ou seja, não é necessário declarar explicitamente o tipo de dado associado a uma variável, como ocorre em outras linguagens como C e Java.

O interpretador do Python se encarrega de identificar automaticamente o tipo de valor que estamos atribuindo a uma variável.

Para saber o tipo de dados de um valor ou variável, podemos perguntar ao Python através do comando type:

````python
print("4 * 2")
type("4 * 2")
````

Como calcular o valor de um produto após um desconto?

Por exemplo, um produto de R\$40 com 20% de desconto sairia por R\$32.

``valor = 40``

``desconto = 20``

O desconto é informado no valor percentual, portanto 20% = 20/100 = 0.02.

O valor descontado é igual ao valor multiplicado pelo desconto, ou seja, 40 x 0.02 = 8
O valor final é igual o valor inicial menos o valor do desconto, ou seja, 40 - 8 = 32.

Como seria um programa? Vejamos a seguir:

````python
# Programa que computa o valor de um produto com desconto
# ENTRADA
valor = input('Digite o valor do produto: ')
valor = float(valor)
desconto = float(input('Digite o desconto, em valores percentuais, para o produto: '))
# PROCESSAMENTO
valor_final = valor - (valor * desconto / 100)
# SAÍDA
print(f'O valor final do produto, após o desconto de {desconto}%, é igual a R${valor_final}')
````

Você gosta de matemática? Não? 😧

Matemática é muito legal pois nos dá atalhos para alguns cálculos.

Vejamos:

Sabemos que se multiplicarmos um número por 1, o resultado é o mesmo: 200 *1 é igual a 200

Sabemos que 0,01 é 1% de qualquer número. Então se subtrairmos 0,01 de 1 teremos 0,99.

Agora se multiplicarmos 0,99 por um número qualquer, nós iremos descontar 1% do valor.

Mágica? Não matemática 🥰

````python
# Programa que computa o valor de um produto com desconto
# ENTRADA
valor = input('Digite o valor do produto: ')
valor = float(valor)
desconto = float(input('Digite o desconto, em valores percentuais, para o produto: '))
# PROCESSAMENTO - Outra forma de fazer o mesmo cálculo
valor_final = valor * (1 - desconto/100)
# SAÍDA
print(f'O valor final do produto, após o desconto de {desconto}%, é igual a R${valor_final}')
````
