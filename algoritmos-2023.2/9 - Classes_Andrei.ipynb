Estruturas (*structs*) ou registros são tipos de dados especiais definidos pelo programador cujo objetivo é armazenar dados relacionados entre si.
Por exemplo: uma estrutura do tipo Pessoa poderia agregar dados como: nome, email, endereço, data nascimento, etc.

Em Python, não temos um comando específico para declarar estruturas. Mas podemos fazer algo semelhante, por meio do conceito de **Classes** (o qual será aprofundado adiante no curso por meio do estudo da Programação Orientada a Objetos).

Exemplo:

```python
# Definindo uma classe com três atributos sem valor padrão
class Carro:
    marca = None
    modelo = None
    ano = None
    cor = None

# Criando um objeto (instância) da classe Carro
meu_carro = Carro()
# Definindo os valores nos atributos
meu_carro.marca = 'VW'
meu_carro.modelo = 'Polo'
meu_carro.ano = 2019
meu_carro.cor = 'Azul'

print("Marca: ", meu_carro.marca)
print("Modelo:", meu_carro.modelo)
print("Ano:", meu_carro.ano)
print("Cor:", meu_carro.cor)

# Criando outra instância da classe Carro
outro_carro = Carro()

outro_carro.marca = 'Chevrolet'
outro_carro.modelo = 'Camaro'
outro_carro.ano = '2022'
outro_carro.cor = 'Amarelo'

print("Marca: ", outro_carro.marca)
print("Modelo:", outro_carro.modelo)
print("Ano:", outro_carro.ano)
print("Cor:", outro_carro.cor)
```

# Criando uma lista com estruturas

```python
class Conta:
    numero = None
    titular = None
    saldo = 0

lista = []
for i in range(3):
    c = Conta()
    c.numero = input("\nNúmero: ")
    c.titular = input("Titular: ")
    c.saldo = input("Saldo: ")
    lista.append(c)

# Exibindo todas as contas
print("\nRelatório de contas:")
for i in range(3):
    print(" numero: ", lista[i].numero, " cliente: ", lista[i].titular, " saldo: ", lista[i].saldo)

# Exibindo uma conta específica
print("\nConta armazenada no índice 2:")
print(f'O cliente {lista[2].titular} possui saldo de R$ {lista[2].saldo}')
```

## Atividades
**3** - Crie uma lista com 3 carros (elementos da classe **Carro**), faça a leitura dos dados e, ao final, imprima os dados dos 3 carros. Baseie-se no código acima para fazer a atividade.

```python
class Carro:
    marca = None
    modelo = None
    ano = None
    cor = None

lista_carros = []
for i in range(3):
    carro = Carro()
    carro.marca = input("Marca: ")
    carro.modelo = input("Modelo: ")
    carro.ano = int(input("Ano: "))
    carro.cor = input("Cor: ")
    lista_carros.append(carro)
    
for i in range(len(lista_carros)):
    print(f"Marca: {lista_carros[i].marca}\nModelo: {lista_carros[i].modelo}\nAno: {lista_carros[i].ano}\nCor: {lista_carros[i].cor}")
```

**4** - Implemente uma função chamada **adiciona_conta** capaz de adicionar uma nova conta (carregado com dados fornecidos pelo usuário) a uma lista.

```python
class Conta:
    numero = None
    titular = None
    saldo = 0
    
def adiciona_conta(numero, titular, saldo):
    conta = Conta()
    conta.numero = numero
    conta.titular = titular
    conta.saldo = saldo
    return conta
conta_luiz = adiciona_conta(123, "luiz", 2)    

print(f"Numero: {conta_luiz.numero}\nTitular: {conta_luiz.titular}\nSaldo: {conta_luiz.saldo}")
```

**5** - Implemente uma função chamada **exibir_contas** capaz de imprimir todas as contas com saldo maior ou igual a N reais. O valor de N deve ser passado por parâmetro para a função assim como a lista de contas.


```python
lista_contas = []
while True:
    inserir_conta = int(input("Digite -1 para parar de inserir contas: "))
    if inserir_conta == -1:
        break
    conta_numero = int(input('Número: '))
    conta_titular = input('Titular: ')
    conta_saldo = float(input('Saldo: '))
    conta = adiciona_conta(conta_numero, conta_titular, conta_saldo)
    lista_contas.append(conta)

def exibir_contas(lista_contas, saldo_minimo):
    for i in range(len(lista_contas)):
        if lista_contas[i].saldo >= saldo_minimo:
            print(f"Numero: {lista_contas[i].numero}\nTitular: {lista_contas[i].titular}\nSaldo: {lista_contas[i].saldo}")
            
exibir_contas(lista_contas, saldo_minimo=30)
```

# Extra: Inicializando valores na classe

É possível inicializar os valores das variáveis da classe no momento da instanciação. Para isso, definimos dentro da classe uma função com nome padrão de `__init__`. Esta função é um construtor: recebe os valores a serem usados na inicialização e atribui cada um à respectiva variável da classe.

A palavra reservada `self` é  usada para mostrar quais variáveis pertencem à classe.

```python
class Carro:
    marca = None
    modelo = None
    ano = None

    def __init__(self, ma, mo, an):
        self.marca = ma
        self.modelo = mo
        self.ano = an

audi = Carro("Audi", "A3", 2023)
print(audi.modelo)
print(audi)

# Com dados via input
carro_usuario = Carro(input("Marca: "), input("Modelo: "), input("Ano: "))
print(carro_usuario.modelo)
```

Podemos facilitar a impressão dos dados de um objeto implementando o método (função) padrão `__str__` na classe.

```python
class Carro:
    marca = None
    modelo = None
    ano = None

    def __init__(self, marca, modelo, ano):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano

    def __str__(self):
        return 'Este carro é da marca ' + str(self.marca) + ', modelo ' + str(self.modelo) + ' e ano ' + str(self.ano)

audi = Carro("Audi", "A3", 2023)
print(audi.modelo)
print(audi)
```
